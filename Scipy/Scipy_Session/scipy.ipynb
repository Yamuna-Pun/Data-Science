```python
import scipy.interpolate as spi
```

```python
import numpy as np
```

```python
# Sample data(
x = np.linspace(0,10,10)     
#Data points for x-axis
                              
y = np.sin(x)
 #correspondig y values
    
#create an interpolation function
interp_func = spi.interp1d(x,y, kind='linear')

#New x values for interpolation 
x_new  = np.linspace(0,10,100)

#Perform interpolation
y_new = interp_func(x_new)

#Plot original data and interpolated curve
plt.figure(figsize=(8,6))
plt.scatter(x,y, label='original data')
plt.plot(x_new,y_new,label='Interpolated curve',color='red')
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('linear Interpolation with Scipy')
```

```python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt

# Generate a noisy signal
t = np.linspace(0, 1, 1000, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(1000)

# Design a low-pass Butterworth filter
order = 4
cutoff_frequency = 10  # Hz
nyquist = 0.5 * 1000  # Nyquist frequency for the signal
normal_cutoff = cutoff_frequency / nyquist
b, a = butter(order, normal_cutoff, btype='low', analog=False)

# Apply the filter to the signal
filtered_signal = filtfilt(b, a, signal)

# Plotting the original and filtered signal
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Low-pass Filtering with SciPy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers from a normal distribution
mean = 0
std_dev = 1
sample_size = 1000

# Generate random numbers from a normal distribution
data = stats.norm.rvs(loc=mean, scale=std_dev, size=sample_size)

# Calculate basic statistics
mean_data = np.mean(data)
std_dev_data = np.std(data)
median_data = np.median(data)

# Create a normal distribution object
normal_dist = stats.norm(mean, std_dev)

# Generate points on x-axis for the probability density function (PDF)
x = np.linspace(-4, 4, 100)
pdf = normal_dist.pdf(x)  # Probability Density Function

# Plotting
plt.figure(figsize=(8, 6))

# Histogram of the generated random numbers
plt.hist(data, bins=30, density=True, alpha=0.5, label='Histogram of Data')

# Plot the probability density function (PDF) of the normal distribution
plt.plot(x, pdf, label='Normal Distribution PDF', color='red')

plt.axvline(mean_data, color='green', linestyle='--', label=f'Mean: {mean_data:.2f}')
plt.axvline(median_data, color='blue', linestyle='--', label=f'Median: {median_data:.2f}')

plt.xlabel('Value')
plt.ylabel('Density/Probability')
plt.title('Normal Distribution and Random Numbers')
plt.legend()
plt.grid(True)
plt.show()
```
